#### Cleaning Datasets and then some EDA

####1.supply_chain dataset

```python
import pandas as pd
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt

supply_df = pd.read_csv(r'C:\Users\harsh\OneDrive\Desktop\data analysis\AI powered supply chain esg and risk optimizatoin\data\supply_chain_full_realistic.csv')

supply_df.head()
```

```python
supply_df.info()
```

```python
supply_df.isnull().sum()
```

```python
supply_df.drop_duplicates()
```

```python
supply_df['Date'] = pd.to_datetime(supply_df['Date'])
```

```python
# Handling missing values
# fill numerical with sector median 
for col in ['Lead_times','Price','Stock_levels','Manufacturing_costs','Shipping_costs']:
    supply_df[col] = supply_df.groupby('Sector')[col].transform(lambda x: x.fillna(x.median()))
```

```python
# fill categorical with mode
for col in ['Shipping_carriers','Transportation_modes','Routes','Customer_demographics']:
    supply_df[col] = supply_df.fillna(supply_df.mode().iloc[0])[col]
```

```python
numeric_col = [col for col in supply_df.columns if supply_df[col].dtype =='float64']
numeric_col
```

```python
# Finding some negative values
(supply_df[numeric_col] <0).sum()
```

```python
# Manufacturing costs has some negative values , fill with 0
supply_df['Manufacturing_costs'] = supply_df['Manufacturing_costs'].fillna(0)
```

## EDA of Supply chain dataset

```python
print("Shape of Dataset",supply_df.shape)
print("Missing values ",supply_df.isnull().sum())
```

```python
# Missingness Heatmap
plt.figure(figsize =(12,6))
sns.heatmap(supply_df.isnull(),cbar = False, cmap = 'viridis')
plt.title("Missing values Heatmap")
plt.show()
```

```python
# Distribution of numerical features
supply_df[['Lead_times','Revenue_generated','Manufacturing_costs','Shipping_costs']].hist(bins=30,figsize=(12,8))
plt.suptitle('Distribution of Key metrices')
plt.show()
```

#### Manufacturing cost , Lead time and revenue generated are right skewed , shipping costs are somewhat normally distributed . we will use log transform for heavily skewed data

```python
# sector wise lead times
plt.figure(figsize=(10,6))
sns.boxplot(x='Sector',y='Lead_times',data= supply_df)
plt.xticks(rotation=45)
plt.title("Sector_wise lead Times")
plt.show()
```

- Energy sector has highest median lead time and also many outliers
- Utilities and Technology sectors have the lowest lead times (shorter and more predictable).
- Financial Services and Communication Services are moderate â€” wider spread but not as extreme as Energy.

```python
# Sector revenue contribution
plt.figure(figsize=(10,5))
sns.barplot(x='Sector',y='Revenue_generated',data=supply_df,estimator=np.sum,ci=None)
plt.xticks(rotation=45)
plt.title('Revenue Contribution by sector')
plt.show()
```

- Energy Sector Contribute more in Revenue. after that Utilities and Technology sectors contribute more 
- From above Lead time distribution , Energy sector has highest lead time and highest revenue - Energy - profitable and operational insufficent
- Utilities & Technology: High revenue + low lead times â†’ efficient and profitable sectors.
- Healthcare: Moderate revenue but higher lead times â†’ could be a bottleneck sector.
- Financial Services & Consumer Cyclical: Low revenue + varying lead times â†’ may not be strong performers.

```python
# Supplier performance Top 10
top_suppliers = supply_df.groupby('Supplier_Name')['Revenue_generated'].sum().nlargest(10)

plt.figure(figsize=(12,6))
sns.barplot(x=top_suppliers.index,y=top_suppliers.values)
plt.xticks(rotation = 45)
plt.title("Top 10 suppliers by Revenue")
plt.show()
```
